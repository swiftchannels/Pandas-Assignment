```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
player_number = len(purchase_data["SN"].unique())
Summary_table =pd.DataFrame([{"Total Players": player_number}])
Summary_table
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Unique_items = len(purchase_data["Item ID"].unique())
revenue = purchase_data["Price"].sum()
purchases = len(purchase_data["Item Name"])
Average = purchase_data["Price"].mean()
Purchase_analysis = [{"Number of Unique Items":Unique_items,
                      "Average Price":Average,
                      "Number of Purchases":purchases,
                    "Total Revenue":revenue}]
Purchase_analysis_df = pd.DataFrame(Purchase_analysis)
purchase_analysis_df = Purchase_analysis_df[['Number of Unique Items', 'Average Price', 'Number of Purchases', 'Total Revenue']]
purchase_analysis_df['Total Revenue'] = purchase_analysis_df['Total Revenue'].map('${:,.2f}'.format)
purchase_analysis_df['Average Price'] = purchase_analysis_df['Average Price'].map('${:,.2f}'.format)
purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [4.99, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 44.99]
label = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

converted_data["Age Range"] = pd.cut(converted_data["Age"], bins, labels=label)
age_range = converted_data.groupby('Age Range').nunique()["SN"]
percentage = age_range / player_number *100

Age_survey_df = pd.DataFrame({"Total Count": age_range,
             "Percentage of Players": percentage})
#Age_survey_df['Percentage of Players'] = Age_survey_df['Percentage of Players'].map('{:,.2f}'.format)
Age_survey_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pur_count = converted_data.groupby('SN').nunique()['Purchase ID']
Averag_purchase_price = converted_data.groupby('SN').mean()['Price']
Tota_purchase_price = Averag_purchase_price * pur_count
Top_spender_df = pd.DataFrame({"Purchase Count": pur_count,"Average Purchase Price": Averag_purchase_price,"Total Purchase Value": Tota_purchase_price})
#Top_spender_df["Average Purchase Price"] = Top_spender_df["Average Purchase Price"].map("${:.2f}".format)
#Top_spender_df["Total Purchase Value"] = Top_spender_df["Total Purchase Value"].map("${:.2f}".format)

Top_spender_df = Top_spender_df.sort_values('Total Purchase Value', ascending=False)
Top_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
arranged_data = converted_data[["Item ID","Item Name","Price"]]

grouped_data = arranged_data.groupby(["Item ID", "Item Name"])

count_purchase = grouped_data["Price"].count()
price_t = grouped_data["Price"].sum()


Most_popular_df = pd.DataFrame({"Purchase Count": count_purchase,
             "Item Price": price_t / count_purchase,
             "Total Purchase Value": price_t})
Most_popular_df['Item Price'] = Most_popular_df["Item Price"].map("${:.2f}".format)
Most_popular_df['Total Purchase Value'] = Most_popular_df["Total Purchase Value"].map("${:.2f}".format)
Most_popular_df = Most_popular_df.sort_values('Purchase Count', ascending=False)
Most_popular_df.head()                                                                                      
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
arranged_data1 = converted_data[["Item ID","Item Name","Price"]]

grouped_data1 = arranged_data1.groupby(["Item ID", "Item Name"])
p_count = grouped_data1["Price"].count()
price_to = grouped_data1["Price"].sum()


profitable_df = pd.DataFrame({"Purchase Count": p_count,
             "Item Price": price_to / p_count,
             "Total Purchase Value": price_to})
#profitable_df["Item Price"] = profitable_df["Item Price"].map("${:.2f}".format)
#profitable_df["Total Purchase Value"] = profitable_df["Total Purchase Value"].map("${:.2f}".format)

profitable_df = profitable_df.sort_values('Total Purchase Value', ascending=False)
profitable_df.head()
```
